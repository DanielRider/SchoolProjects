```python
import numpy as np
#Goal is to generate 500 points on the surface of a 50-dimensional ball
```

```python
d = 50
n = 500
mu = 0.0
sigma = 1.0
#mean of 0 and unit variance 
```

```python
A = [[np.random.default_rng().normal(mu,sigma,d)] for i in range(n)]
A = np.squeeze(np.asarray(A))
#This generates a 50 x 500 array where each row contains 50 points generated from a gaussian distribution
#These rows will be our coordinates in the various dimensions.
#Then we create 500 rows of the same type.  
#Now we want to normalize the rows so that they are on the surface of our ball
for i in range(n):
  A[i] =  A[i] / np.linalg.norm(A[i])
```

```python
#At this point is when I gave up last night and now im back with a method to do the hemisphere stuff
#My idea in my head is using a two hyperplanes to split the sphere and then we can determine if a point is inside
#or outside the two hyperplanes using absolute value.
#There is probably a simpler explanation for what im actually doing but thats what i thought up
#PLAN:
#"Create" two hyperplanes centered around the origin of the equation w dot x = b
#The distance of each hyper plane from the origin is (1/sqrt(50)) which is our b  
#A point (x) is inside the band region if abs(w dot x) < (1/sqrt(50)
```

```python
W = [[np.random.default_rng().normal(mu,sigma,d)] for i in range(5)]
for i in range(5):
  W[i] =  W[i] / np.linalg.norm(W[i])
W = np.squeeze(np.asarray(W))
#generating 5 new points
```

```python
#W will be the matrix with our 5 newly generated points

b = 1/np.sqrt(50)
inside = [0 for i in range(5)]
allfive = [0 for i in range(n)]

c = 0

for i in range(5):        #for every generated point on surface of ball
  for j in range(n):      #for every band
    if(np.absolute(np.dot(W[i],A[j])) <= b):       #this line is pretty much my entire idea. Is the abs(dotproduct(w,a)) < b?
      inside[i]+=1
      allfive[j]+=1

for i in allfive:
  if i == 5:       #side code for other questions
    c+=1

print(inside)
print("There are ")
print(c)
print( "points in all five bands.")
#inside gives us the number of points in each region
#c gives us the number of points in all bands
#we need b=.485. Since b is the distance from the origin or band width.  The size of the band is nearly the entire sphere.
```
